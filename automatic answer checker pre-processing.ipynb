```python
regex=nltk.RegexpTokenizer(r'\w+') #tokenizing and removing punctuations
df['Text Tokenize']=df['EssayText'].apply(lambda x:regex.tokenize(x))
```

```python
def stopword(tokenized_words):  #removing stop words
    t=[w for w in tokenized_words if w not in stopwords]
    return t
df['stopwords removal']=df['Text Tokenize'].apply(lambda x:stopword(x))
```

```python
l = nltk.WordNetLemmatizer()
t=['Some', 'additional', 'information', 'would','reading']
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
def get_wordnet_pos(word):      #function to tag every word in the text with POS
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
lem=[lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in t]#lemmaztizing every word in the text with POS using function
def lemmatization(text):
    return [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in text]
df['lemmatization']=df['stopwords removal'].apply(lambda x:lemmatization(x))
```
